```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
num_stars = 1000  # Galaksi için kullanılacak yıldız sayısı
disk_radius = 10  # Galaksi diskinin yarıçapı
disk_thickness = 0.2  # Galaksi diskinin kalınlığı
disk_mass = 1000  # Galaksi diskinin kütlesi
```

```python
theta = np.random.uniform(0, 2*np.pi, num_stars)
radius = np.random.uniform(0, disk_radius, num_stars)
x = radius * np.cos(theta)
y = radius * np.sin(theta)
```

```python
v_circular = np.sqrt(disk_mass * disk_radius / num_stars)  # Yıldızların dönme hızı
v = v_circular * np.sqrt(radius)  # Yıldızların rastgele hızları
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(x, y, c='b', s=2)  # Yıldızları çizdirme
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(-disk_radius, disk_radius)
plt.ylim(-disk_radius, disk_radius)
plt.title('Galaksi Oluşumu Simülasyonu')
plt.show()
```

```python
dt = 0.01  # Zaman adımı
num_steps = 100  # Simülasyon adım sayısı
```

```python
for step in range(num_steps):
    # Konumları güncelleme
    x += v * np.cos(theta) * dt
    y += v * np.sin(theta) * dt
```

```python
theta = np.arctan2(y, x)
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(x, y, c='b', s=2)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-disk_radius, disk_radius)
plt.ylim(-disk_radius, disk_radius)
plt.title(f'Galaksi Oluşumu Simülasyonu - Adım {step+1}')
plt.show()
```

```python
X_train = np.linspace(0, 10, 100)
y_train = 2*X_train + 1
```

```python
X_test = np.linspace(0, 10, 20)
y_test = 2*X_test + 1
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# Modeli oluşturma
model = LinearRegression()
```

```python
# Modeli eğitme
model.fit(X_train.reshape(-1, 1), y_train)
```

```python
# Tahminler yapma
y_pred = model.predict(X_test.reshape(-1, 1))
```

```python
from sklearn.metrics import mean_squared_error
```

```python
# MSE hesaplama
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```
